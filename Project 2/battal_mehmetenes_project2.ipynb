Naive Bayes makes the assumption that  x0  through  xn  are conditionally independent given  ci , which allows us to write

P(x0,…,xn|ci)=P(x0|ci)∗P(x1|ci)∗…∗P(xn|ci). And according to the Bayes' theorem p(c|x) = p(x|c)*p(c)/p(x). We will use these in our calculations. Also we omit p(x) since it would be same for each class label.


In this question we need the probabilities of:
P(Red|Yes), P(SUV|Yes), P(Imported|Yes), P(Red|No) , P(SUV|No), P(Imported|No) and P(Yes), P(No) to multiply with the respective probabilities.


From the data we can calculate these probibilities:

P(Red|Yes) = 3/5; P(SUV|Yes) = 1/5; P(Imported|Yes) = 3/5; P(Red|No) = 2/5

P(SUV|No) = 3/5; P(Imported|No) = 2/5; p(yes) = 1/2; p(No) = 1/2


Since we are trying to find if it is stolen:

p(yes|red, suv, imported) = P(Red|Yes)*P(SUV|Yes)*P(Imported|Yes)*p(yes)

p(yes|red, suv, imported) = 3/5 * 1/5 * 3/5 * 1/2 = 9/250 = 0.036

p(no|red, suv, imported) = P(Red|No) * P(SUV|No) * P(Imported|No) * p(no)

p(no|red, suv, imported) = 2/5 * 3/5 * 2/5 * 1/2 = 12/250 = 0.048

Since the probability of not getting stolen is bigger than getting stolen (0.048 > 0.036), Naive Bayes would predict the label "No" with the given attributes color=red, type=suv and origin=imported.

3. Display the resulting classification boundaries in a **1x3 layout**. Utilize the axes titles to highlight the models.

  - Here, you should **put the train data as a parameter to the `visualize_model` function**, i.e. the function call should be visualize_model(model, x_train, y_train, ax, title).

```
from matplotlib.colors import ListedColormap
from ipywidgets import interact

def visualize_model(model, data, labels, ax, title):
  step = 0.05

  cmap_light = ListedColormap(['orange', 'cyan', 'cornflowerblue'])
  cmap_bold = ListedColormap(['darkorange', 'c', 'darkblue'])

  model.fit(data, labels)

  x_min = data[:, 0].min() - 1
  x_max = data[:, 0].max() + 1
  y_min = data[:, 1].min() - 1
  y_max = data[:, 1].max() + 1

  xx, yy = np.meshgrid(np.arange(x_min, x_max, step), np.arange(y_min, y_max, step))
  y_pred = model.predict(np.hstack([xx.reshape(-1,1), yy.reshape(-1,1)])).reshape(xx.shape)

  ax.pcolormesh(xx, yy, y_pred, cmap=cmap_light)
  ax.scatter(data[:, 0], data[:, 1], c=labels, cmap=cmap_bold, edgecolor='k', s=20)
  ax.set_xlim(xx.min(), xx.max())
  ax.set_ylim(yy.min(), yy.max())
  ax.set_title(title)
```

```
from sklearn.datasets import make_classification

features, labels = make_classification(n_samples=100, 
                    n_classes=2,
                    n_features=2, 
                    n_informative=2,
                    class_sep=0.5,
                    n_redundant=0)

plt.scatter(features[:,0], features[:,1], c=labels)
plt.grid()
plt.show()
```

```
# your code
from sklearn import tree

X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.20, random_state=1)

model1 = tree.DecisionTreeClassifier(min_samples_leaf = 1)
model2 = tree.DecisionTreeClassifier(min_samples_leaf = 5, max_depth = 4)
model3 = tree.DecisionTreeClassifier(min_samples_leaf = 20, max_depth = 4)

fig, ax = plt.subplots(1, 3, figsize=(14,7))

visualize_model(model1, X_train, y_train, ax[0], "Model 1\n(min_samples_leaf = 1)")
visualize_model(model2, X_train, y_train, ax[1], "Model 2\n(min_samples_leaf = 5, max_depth = 4)")
visualize_model(model3, X_train, y_train, ax[2], "Model 3\n(min_samples_leaf = 20, max_depth = 4)")
```
